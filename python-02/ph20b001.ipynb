```python
def conway(init: np.ndarray, niter:int)->np.ndarray:
    
    A = init.copy()
    N = np.shape(A)[0]
    ans_array = np.zeros(np.shape(A),dtype=int)
    for k in range(niter):
        for r in range(1,N-1):
            for c in range(1,N-1):
                       
        
        
                if A[r][c]==0:
                    f=0
                    for i in range(3):
                        for j in range(3):
                            f =f+A[r+i-1][c+j-1]
                    #print(f)
                    if(f==3):
                        ans_array[r][c]=1
                    else:
                        ans_array[r][c]=0
                else:
                    f=0
                    for i in range(3):
                        for j in range(3):
                            f = f + A[r + i - 1][c + j - 1]
                    #print(f)
                    f= f-1
                    if(f==2) or (f==3):
                        ans_array[r][c]=1
                    else:
                        ans_array[r][c]=0
        A=ans_array.copy()
    return ans_array    
```
