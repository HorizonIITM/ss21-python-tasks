```python
def conway(init:np.ndarray, niter:int) -> np.ndarray:
    m=np.size(init,0)
    n=np.size(init,1)
    count=0
    for k in range(0,niter):
        for i in range(1,m-1):
            for j in range(1,n-1):
                count=0
                if(init[i-1][j-1]==1): count+=1
                if(init[i-1][j]==1): count+=1
                if(init[i-1][j+1]==1): count+=1
                if(init[i][j-1]==1): count+=1
                if(init[i][j+1]==1): count+=1
                if(init[i+1][j-1]==1): count+=1
                if(init[i+1][j]==1): count+=1
                if(init[i+1][j+1]==1): count+=1
                if(init[i][j]==1):
                    if(count==2 or count==3): init[i][j]=1
                    else: init[i][j]=0
                else:
                    if(count==3): init[i][j]=1
                    else: init[i][j]=0
    return init

plt.imshow(test_array, cmap='binary')
final_array=conway(test_array,1337)
plt.imshow(final_array, cmap='binary', alpha=0.9)
plt.box(False)
plt.axis('off')
plt.savefig('img/ae20b058.png', bbox_inches='tight',pad_inches = 0)
        
```
