```python
def conway(init, n):
    import numpy as np
    k = len(init[0,:])
    j = len(init[:,0])
    arr = np.zeros([k,j], dtype = int)
    new_arr = np.zeros([k+2,j+2], dtype = int)
    new_arr[1:k+1,1:j+1] = init
    for i in range(n):
        for m in range(1, k+1):
            for o in range(1, j+1):
                r = 0 
                for p in range(-1,2):
                    for q in range(-1,2):
                        if p != 0 and q != 0:
                            r+= new_arr[m+p,o+q]
                        else:
                            continue
                if new_arr[m,o]==1:
                    if r in range(2,4):
                        arr[m-1,o-1] = 1
                    else:
                        arr[m-1,o-1] = 0
                if new_arr[m,o]==0:
                    if r==3:
                        arr[m-1,o-1] = 1
                    else:
                        arr[m-1,o-1] = 0
    return arr           
```
