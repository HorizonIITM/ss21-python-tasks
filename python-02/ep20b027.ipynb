```python
def conway(init, niter = 1):
    m, n = np.shape(init)
    
    ext_init = np.zeros((m + 2, n + 2))
    ext_init[1 : m + 1, 1 : n + 1] = init
    
    output = np.zeros((m, n))
    output[:, :] = init
    
    for iteration in range(niter):
        for i in range(1, m + 1):
            for j in range(1, n + 1):
                sum = ext_init[i - 1][j] + ext_init[i + 1][j] + ext_init[i][j - 1] + ext_init[i][j + 1]
                elem = ext_init[i][j]
                
                if sum in (2, 3) and elem == 1:
                    output[i - 1][j - 1] = 1
                    
                elif sum == 3 and elem == 0:
                    output[i - 1][j - 1] = 1
                    
                else:
                    output[i - 1][j - 1] = 0
            
        ext_init[1 : m + 1, 1 : n + 1] = output
    
    return output
```
