```python
def conway(init, niter):
    r = init.shape[0]
    c = init.shape[1]
    k=0
    while( np.all(init == 0) or k < niter):
        for i in range(r):
            for j in range(c):
                if( (i>0 and i<r-1) and (j>0 and j<c-1) ):
                    sum = init[i-1][j]+init[i+1][j]+init[i][j-1]+init[i][j+1]+init[i-1][j-1]+init[i+1][j+1]+init[i-1][j+1]+init[i+1][j-1]
                    if(init[i][j]==0):
                        if(sum == 3):
                            init[i][j] = 1
                        else:
                            init[i][j] = 0
                    else:
                        if((sum == 2) or (sum == 3)):
                            init[i][j] = 1
                        else:
                            init[i][j] = 0
                elif( ( j>0 and j<c-1) and (i==0) ):
                    sum = init[i][j-1]+init[i][j+1]+init[i+1][j+1]+init[i+1][j]+init[i+1][j-1]
                    if(init[i][j]==0):
                        if(sum == 3):
                            init[i][j] = 1
                        else:
                            init[i][j] = 0
                    else:
                        if((sum == 2) or (sum == 3)):
                            init[i][j] = 1
                        else:
                            init[i][j] = 0
                elif( (j>0 and j<c-1) and i==r-1):
                    sum = init[i][j-1]+init[i][j+1]+init[i-1][j-1]+init[i-1][j]+init[i-1][j+1]
                    if(init[i][j]==0):
                        if(sum == 3):
                            init[i][j] = 1
                        else:
                            init[i][j] = 0
                    else:
                        if((sum == 2) or (sum == 3)):
                            init[i][j] = 1
                        else:
                            init[i][j] = 0
                elif( (i==0) and (j==0) ):
                    sum = init[0][1]+init[1][0]+init[1][1]
                    if(init[i][j]==0):
                        if(sum == 3):
                            init[i][j] = 1
                        else:
                            init[i][j] = 0
                    else:
                        if((sum == 2) or (sum == 3)):
                            init[i][j] = 1
                        else:
                            init[i][j] = 0
                elif( (i==0) and (j==c-1) ):
                    sum = init[0][c-2]+init[1][c-2]+init[1][c-1]
                    if(init[i][j]==0):
                        if(sum == 3):
                            init[i][j] = 1
                        else:
                            init[i][j] = 0
                    else:
                        if((sum == 2) or (sum == 3)):
                            init[i][j] = 1
                        else:
                            init[i][j] = 0
                elif( (i==r-1) and (j==0) ):
                    sum = init[r-1][1]+init[r-2][0]+init[r-2][1]
                    if(init[i][j]==0):
                        if(sum == 3):
                            init[i][j] = 1
                        else:
                            init[i][j] = 0
                    else:
                        if((sum == 2) or (sum == 3)):
                            init[i][j] = 1
                        else:
                            init[i][j] = 0
                elif( (i==r-1) and (j==c-1) ):
                    sum = init[r-1][c-2]+init[r-2][c-2]+init[r-2][c-1]
                    if(init[i][j]==0):
                        if(sum == 3):
                            init[i][j] = 1
                        else:
                            init[i][j] = 0
                    else:
                        if((sum == 2) or (sum == 3)):
                            init[i][j] = 1
                        else:
                            init[i][j] = 0
        k = k + 1
    return init
```
