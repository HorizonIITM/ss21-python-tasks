```python
def conway(init: np.ndarray, niter: int) -> np.ndarray:
    iter_array=np.zeros(init.shape)
    for it in range(niter-1):
        for i in range(init.shape[0]-1):
            for j in range(init.shape[1]-1):
                if i!=0 or j!=0 or i!=init.shape[0] or j!= init.shape[1]:
                    if init[i][j]==1:
                        if (init[i-1][j]+init[i][j-1]+init[i][j+1]+init[i+1][j])==2 or (init[i-1][j]+init[i][j-1]+init[i][j+1]+init[i+1][j])==3:
                            iter_array[i][j]=1
                    else:
                        if (init[i-1][j]+init[i][j-1]+init[i][j+1]+init[i+1][j])==3:
                            iter_array[i][j]=1
        init=iter_array
    return iter_array
```
