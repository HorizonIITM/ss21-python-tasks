```python
def conway(init: np.ndarray, niter: int) -> np.ndarray:
    arr_new = init.copy()
    x,y = init.shape
    if (niter == 0):
        return init
    else:
        niter -= 1
        
    for i in range(1,x-1):
        for j in range(1,y-1):
            sum = init[i-1,j+1] + init[i,j+1] + init[i+1,j+1] + init[i+1,j] + init[i+1,j-1] + init[i,j-1] + init[i-1,j-1] + init[i-1,j]
            if (init[i,j] == 1):
                if (sum == 2):
                    arr_new[i,j] = 1
                elif (sum == 3):
                    arr_new[i,j] = 1
                else:
                    arr_new[i,j] = 0
            else:
                if (sum == 3):
                    arr_new[i,j] = 1
    
    return conway(arr_new, niter)

iter_array = conway(test_array,1337)
plt.figure()
plt.imshow(test_array, cmap='binary')
plt.imshow(iter_array, cmap='binary', alpha=0.9)
plt.box(False)
plt.axis('off')
plt.savefig('img/mm20b001.png', bbox_inches='tight', pad_inches=0)
```
